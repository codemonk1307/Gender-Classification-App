```python
from PIL import Image # Python Image Library
import matplotlib
import cv2 as cv
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading image using PIL
img = Image.open("disappointed.jpg")
img
```

```python
# Loading images using matplotlib   by-default reads  -->  RGB 
img_mat = matplotlib.pyplot.imread("disappointed.jpg")
img_mat
```

```python
# Loading images using openCV by-default openCV reads --> BGR 
img_cv = cv.imread("disappointed.jpg")
img_cv
```

```python
# (Rows, Columns, Depth)
# by-default openCV reads --> BGR 
img_cv.shape
```

```python
# rendering/plotting openCV array  (BGR)
plt.imshow(img_cv)
```

```python
# rendering/plotting matplotlib array  (RGB)
plt.imshow(img_mat)
```
