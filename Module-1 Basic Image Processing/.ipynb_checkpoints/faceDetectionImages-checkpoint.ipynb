```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Step -1 read image 
img = cv.imread("profile photo sachin.jpg")
# Step -2 convert into gray scale image
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# Step -3 applying the haar Cascade Classifier
haar = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
# (gray, scale size, minimum neighbours)
faces = haar.detectMultiScale(gray, 1.3, 5)
```

```python
# [[x-cord, y-cord, width, height]]
print(faces)
```

```python
cv.rectangle(img, (353, 218), (353 + 469, 218 + 469), (0, 255, 0), 7)
plt.imshow(img)
```

```python
# to display image in a pop-up window
cv.imshow("object_detect", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# We can also crop & save the image using slicing
face_crop = img[218: 218 + 469, 353: 353 + 469]
plt.imshow(face_crop)
```

```python
cv.imwrite("profileFaceCrop.jpg", face_crop)
```
